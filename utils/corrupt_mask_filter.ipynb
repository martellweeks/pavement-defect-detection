```python
from sklearn.model_selection import train_test_split
import pandas as pd
from pycocotools.coco import COCO
from glob import glob
import shutil
import os
from tqdm import tqdm
import json
```

```python
coco_file_path = "../data/annotations/val.json"
coco_json_file = open(coco_file_path)
coco_json = json.load(coco_json_file)
```

```python
for anno in coco_json['annotations']:
    print(anno['segmentation'])
print(len(coco_json['annotations']))
```

```python
valid_annotations = list()
for anno in coco_json['annotations']:
    if type(anno['segmentation']) != list:
        continue
    elif len(anno['segmentation']) == 0:
        continue
    elif len(anno['segmentation'][0]) == 0:
        continue
    valid_annotations.append(anno)
```

```python
print(len(valid_annotations))
```

```python
coco_json['annotations'] = valid_annotations
```

```python
for anno in valid_annotations:
    print(anno['segmentation'])
```

```python
with open("../data/annotations/val_filtered.json", "w") as outfile:
    json.dump(coco_json, outfile)
```
